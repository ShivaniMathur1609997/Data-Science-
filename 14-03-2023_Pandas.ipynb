# Missing Values
1. Drop the null values

2. Replace them with the mean, median , and mode - technique -imputation

```python
movie = pd.read_csv('http://bit.ly//imdbratings')
movie
```

```python
movie.isnull().sum()
```

```python
#With the axis you can delete the whole row and column
movie = movie.dropna(axis = 0)
movie
```

```python
movie.isnull().sum()
```

```python
#imputation
movie.isnull().sum()
```

```python
movies1= movie['content_rating']
m1 = movies1.mode()
print(m1)
movie = movies1.fillna(m1,inplace = True)
```
