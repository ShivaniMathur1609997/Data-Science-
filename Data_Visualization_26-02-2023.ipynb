Question 1 Create scatter plots for the following data, make sure all the plots appear in the same plane with x-labels and y-labels evenly spaced.

1.‘hp’ and ‘mpg’.

2.‘hp’ and ‘time-to-60’.

3.‘hp’ and ‘weightlbs’.

4.‘hp’ and ‘cylinders’.

5.Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg, hp-time-to-60) with respect to each brand.

```python
fig,axes=plt.subplots(2,2,figsize=(10,5))

plt.subplot(2,2,1)
sns.scatterplot(x=data['hp'],y=data['mpg'],data=data)
plt.title(' HP data to MPG data')
plt.ylabel('Miles per Gallon')
plt.xlabel('Horsepower')

plt.subplot(2,2,2)
sns.scatterplot(x=data['hp'],y=data['time-to-60'],data=data)
plt.title('HP to time-to-60 Data')
plt.ylabel('Time-to-60')
plt.xlabel('Horsepower')

plt.subplot(2,2,3)
sns.scatterplot(x=data['hp'],y=data['weightlbs'],data=data)
plt.title('HP to Weight data')
plt.ylabel('Weightlbs')
plt.xlabel('Horsepower')

plt.subplot(2,2,4)
sns.scatterplot(x=data['hp'],y=data['cylinders'],data=data)
plt.title('HP data to Cylinders')
plt.ylabel('Cylinders')
plt.xlabel('Horsepower')

plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

```python
fig,axes=plt.subplots(2,2,figsize=(20,10))

plt.subplot(2,2,1)
sns.scatterplot(x=data['hp'],y=data['mpg'],hue=data['brand'],data=data)
plt.title('MPG to HP data')
plt.ylabel('Miles per Gallon')
plt.xlabel('Horsepower')

plt.subplot(2,2,2)
sns.scatterplot(x=data['hp'],y=data['time-to-60'],hue=data['brand'],data=data)
plt.title('HP to time-to-60 Data')
plt.ylabel('Time-to-60')
plt.xlabel('Horsepower')

plt.subplot(2,2,3)
sns.scatterplot(x=data['mpg'],y=data['weightlbs'],hue=data['brand'],data=data)
plt.title('MPG to Weight data')
plt.ylabel('Weightlbs')
plt.xlabel('Horsepower')

plt.subplot(2,2,4)
sns.scatterplot(x=data['hp'],y=data['cylinders'],hue=data['brand'],data=data)
plt.title('Cylinders to HP data')
plt.ylabel('Cylinders')
plt.xlabel('Horsepower')
```

Question 2 Create bar plots for the following, make sure all the plots appear in the same plane with x-labels and y-labels evenly spaced.

1.Cylinders and Horsepower, Cylinders and miles per gallon, Cylinders and weight.

2.Year and time-to-60, Year and miles per gallon, Year and horsepower.

3.Create a bar plot that shows the visual representation of hp, mpg, weight, time-to-60 with respect to the number of cylinders for each of the brands.

4.Create a bar plot that shows the visual representation of hp, mpg, weight and time-to-60 with respect to the years, for each brand.

```python
fig,axes=plt.subplots(3,2,figsize=(15,10))

plt.subplot(3,2,1)
sns.barplot(x=data['cylinders'],y=data['hp'],data=data)
plt.title('Cylinder to HP data')
plt.xlabel('Cylinder')
plt.ylabel('Horsepower')

plt.subplot(3,2,2)
sns.barplot(x=data['cylinders'],y=data['mpg'],data=data)
plt.title('Cylinders to MPG Data')
plt.xlabel('Cylinders')
plt.ylabel('MPG')

plt.subplot(3,2,3)
sns.barplot(x=data['cylinders'],y=data['weightlbs'],data=data)
plt.title('Cylinders to Weight data')
plt.ylabel('Weightlbs')
plt.xlabel('Cylinders')

plt.subplot(3,2,4)
sns.barplot(x=data['year'],y=data['time-to-60'],data=data)
plt.title('Year to time-to-60 data')
plt.ylabel('time-to-60')
plt.xlabel('Year')

plt.subplot(3,2,5)
sns.barplot(x=data['year'],y=data['mpg'],data=data)
plt.title('Year to mpg data')
plt.ylabel('MPG')
plt.xlabel('Year')

plt.subplot(3,2,6)
sns.barplot(x=data['year'],y=data['hp'],data=data)
plt.title('Year to HP data')
plt.ylabel('Hp')
plt.xlabel('Year')

plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

```python
xes=plt.subplots(2,2,figsize=(20,10))

plt.subplot(2,2,1)
sns.barplot(x=data['cylinders'],y=data['hp'],hue=data['brand'],data=data)
plt.title('Cylinder to HP data')
plt.xlabel('Cylinder')
plt.ylabel('Horsepower')

plt.subplot(2,2,2)
sns.barplot(x=data['cylinders'],y=data['mpg'],hue=data['brand'],data=data)
plt.title('Cylinders to MPG Data')
plt.xlabel('Cylinders')
plt.ylabel('MPG')

plt.subplot(2,2,3)
sns.barplot(x=data['cylinders'],y=data['weightlbs'],hue=data['brand'],data=data)
plt.title('Cylinders to Weight data')
plt.ylabel('Weightlbs')
plt.xlabel('Cylinders')

plt.subplot(2,2,4)
sns.barplot(x=data['cylinders'],y=data['time-to-60'],hue=data['brand'],data=data)
plt.title('Cylinders to Time-to-60 data')
plt.ylabel('Time-to-60')
plt.xlabel('Cylinders')

plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()
```

Create a histogram for the following:
