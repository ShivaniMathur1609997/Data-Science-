Data Preprocessing

```python
encod_col = [ 'fueltypes', 'aspiration', 'doornumbers',
       'carbody', 'drivewheels', 'enginelocation','enginetype', 'cylindernumber',
       'fuelsystem']
```

```python
data.nunique()
```

```python
#this code will make a dictionary which will represent which value got which encoding value
encod_dict = {}
for i in encod_col:
    ulist = list(data[i].unique())
    incod_dict = {}
    for n, u in enumerate(ulist):
        incod_dict[u] = n
    encod_dict[i] = incod_dict
```

```python

encod_dict
```

```python
datac = data.copy()

for i in encod_dict:
    datac[i] = datac[i].map(encod_dict[i])
datac.head(20)
```

```python
ndata = datac.drop(["ID","name"], axis=1)
```

```python
#Seperated target and features
x = pd.DataFrame(ndata.iloc[:,:-1])
y = pd.DataFrame(ndata.iloc[:,-1])
```

```python
# Functio to Calculate Adjusted R-Square
def Adj_Rsquare (model , X_train , y_train):
    RSquare = model.score( X_train, y_train)
    N = X_train.shape[0]
    K = X_train.shape[1]
    Adj_Rsquare_value = 1-((1-RSquare) * (N-1) / (N-K-1))
    return Adj_Rsquare_value
```

```python
Adj_R2 = Adj_Rsquare(linreg,x_train , y_train)
print("Adjusted R-Square : ",Adj_R2)
```

R-squared (R2) score and adjusted R-squared (R2 adj) score are both used to evaluate the performance of regression models. R-squared is a statistical measure that represents the proportion of the variance in the dependent variable that is explained by the independent variable(s) in a regression model. R-squared ranges from 0 to 1, with higher values indicating better performance. However, as the number of independent variables increases, the R-squared value tends to increase, even if the added variables do not significantly improve the model. Therefore, it is not a reliable indicator of the model's quality when comparing models with different numbers of independent variables.

Adjusted R-squared, on the other hand, adjusts the R-squared value to account for the number of independent variables in the model. It penalizes the addition of unnecessary independent variables that do not improve the model's performance. Adjusted R-squared ranges from -1 to 1, with higher values indicating better performance. Unlike R-squared, a higher adjusted R-squared value is always better than a lower value when comparing models with different numbers of independent variables.

In summary, R-squared is a simple and commonly used metric for evaluating the goodness of fit of a regression model, while adjusted R-squared is a modified version of R-squared that accounts for the number of independent variables in the model and is therefore more reliable when comparing models with different numbers of independent variables.
